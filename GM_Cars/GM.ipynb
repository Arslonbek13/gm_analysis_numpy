```python
import numpy as np

def load_data(file_name):
    data = np.load(file_name)
    models = data['models']
    qnts = data['qnts']
    return models, qnts
# 2020 yilgi ma'lumotlarni yuklash
models_2020, qnts_2020 = load_data('GM_Cars/gm_2020.npz')

# 2021 yilgi ma'lumotlarni yuklash
models_2021, qnts_2021 = load_data('GM_Cars/gm_2021.npz')
models_2020, qnts_2020
```

```python
def total_qnts():
    qnts_2020
    qnts_2021
    global gm2020, gm2021
    gm2020 = np.sum(qnts_2020)
    gm2021 = np.sum(qnts_2021)
    return gm2020, gm2021
total_qnts()
```

```python
def sel_models():
    sum_2020 = (models_2020 == 'Damas') | (models_2020 == 'Nexia') | (models_2020 == 'Cobalt') | (models_2020 == 'Gentra')
    sum_2021 = (models_2021 == 'Damas') | (models_2021 == 'Nexia') | (models_2021 == 'Cobalt') | (models_2021 == 'Gentra')
    global four_2020, four_2021
    four_2020= np.sum(qnts_2020[sum_2020])
    four_2021 = np.sum(qnts_2021[sum_2021])
    return four_2020, four_2021
sel_models()
```

```python
def ratio_sel_models():
    ratio_2020 = (four_2020/gm2020)*100
    ratio_2021 = (four_2021/gm2021)*100
    return ratio_2020, ratio_2021
ratio_sel_models()
```
